# Random Forest Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset=pd.read_csv("preprocessed_weather.csv")
X=dataset.iloc[:,[i for i in range(13)if i!=7] ].values
Y=dataset.iloc[:,7:8].values
```

Spliit data into training and test

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.25, random_state=42)
```

## Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestRegressor
Reggressor=RandomForestRegressor(n_estimators=10, random_state=42)
Reggressor.fit(x_train,y_train)
```

## Predicting the test set

```
y_pred=Reggressor.predict(x_test)
```

```
x_pred=Reggressor.predict(x_train)
```

## Evaluate the model using MSE and R square error

```
from sklearn.metrics import mean_squared_error, r2_score
mse=mean_squared_error(y_test, y_pred)
r2= r2_score(y_test,y_pred)
```

```
mse_x=mean_squared_error(y_train,x_pred)
```

```
r2_x= r2_score(y_train,x_pred)
```

```
print("Mean Squared Error for y_pred and y_test:", mse)
print("R-squared y_pred and y_test:", r2)
```

```
print("Mean Squared Error for x_train y_train:", mse_x)
print("R-squared x_train and y_train:", r2_x)
```
