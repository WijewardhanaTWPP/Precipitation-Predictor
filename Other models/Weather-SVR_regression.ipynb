Import Library

```python
import numpy as np
import pandas as pd
```

Importing the dataset

```python
dataset=pd.read_csv("preprocessed_weather.csv")
X=dataset.iloc[:,[i for i in range(13)if i!=7] ].values
Y=dataset.iloc[:,7:8].values
```

```python
Y = Y.ravel()
```

Spliit the data in to test and train test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.25, random_state=42)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

Training the SVR model

```python
from sklearn.svm import SVR
svr_model = SVR(kernel='rbf')  # 'rbf' stands for Radial basis function (Gaussian)
svr_model.fit(x_train_scaled, y_train)
```

Predict the test set

```python
y_pred=svr_model.predict(x_test)
```

Predict the train

```python
x_pred=svr_model.predict(x_train)
```

Evaluate the model using MSE and R square error

```python
from sklearn.metrics import mean_squared_error, r2_score
test_mse=mean_squared_error(y_test,y_pred)
test_r2=r2_score(y_test,y_pred)
```

```python
train_mse=mean_squared_error(y_train,x_pred)
train_r2=r2_score(y_train,x_pred)
```

```python
print("Mean Squared Error for test:", test_mse)
print("R-squared for test:", test_r2)
```

```python
print("Mean Squared Error for train:", train_mse)
print("R-squared for train:", train_r2)
```
